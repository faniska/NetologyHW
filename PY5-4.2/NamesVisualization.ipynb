```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
DATA_PATH = '/Users/fanis/AnacondaProjects/Netology-PY5/names/'
```

# Part 1

```python
names_by_year = {}
for year in range(1900, 2000):
    file_path = '{}yob{}.txt'.format(DATA_PATH, year)
    names_by_year[year] = pd.read_csv(file_path, names=['Name','Gender','Count'])
    
name_dynamics = pd.concat(names_by_year).query('Name == ["Ruth","Robert"]')
data = name_dynamics.groupby([name_dynamics.index.get_level_values(0), 'Name']).sum().unstack('Name')
data.plot()
```
